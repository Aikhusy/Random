```python
import numpy as np
import pandas as pd
from sklearn.tree import DecisionTreeClassifier # import DT
from sklearn.ensemble import RandomForestClassifier # import RandomForest
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report
```

1. Mushroom

```python
#membaca data
df=pd.read_csv("mushrooms.csv")
df.head()
```

```python
df.isnull().sum()
```

```python
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()

# Loop melalui semua kolom dalam DataFrame
for column in df.columns:
    # Periksa apakah tipe data kolom adalah objek (data kategorikal)
    if df[column].dtype == 'object':
        df[column] = label_encoder.fit_transform(df[column])
```

```python
df.head()
```

```python
X = df.iloc[:,2:]
y = df['class']
```

```python
X.head()
```

```python

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)
```

```python
decision_tree = DecisionTreeClassifier()
```

```python
param_grid_decision_tree = {
    'criterion': ['gini', 'entropy'],
    'max_depth': [None, 10, 20, 30],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4]
}
```

```python
from sklearn.model_selection import train_test_split, GridSearchCV
grid_search_decision_tree = GridSearchCV(decision_tree, param_grid_decision_tree, cv=5, scoring='accuracy')
grid_search_decision_tree.fit(X_train, y_train)
```

```python
best_params_decision_tree = grid_search_decision_tree.best_params_
best_accuracy_decision_tree = grid_search_decision_tree.best_score_
print("Best parameters for Decision Tree:", best_params_decision_tree)
print("Best accuracy for Decision Tree:", best_accuracy_decision_tree)
```
