```python
path = '/content/drive/MyDrive/SPK/Data WSM dan WPM.csv'
df=pd.read_csv(path)

df.head()
```

```python
kosong = pd.DataFrame(columns=df.columns)

panjangKolom = len(df['c1'])
x=0
while x < panjangKolom:
    kosong.at[x, 'c1'] = df.at[x, 'c1'] / df['c1'].max()
    x += 1

kosong.head()

```

```python
def normalisasiBenefit(DATAFRAME_KOSONG,NAMA_KOLOM,DATAFRAME_ASLI):
  panjangKolom = len(df[NAMA_KOLOM])
  x=0
  while x < panjangKolom:
      DATAFRAME_KOSONG.at[x, NAMA_KOLOM] = DATAFRAME_ASLI.at[x, NAMA_KOLOM] / DATAFRAME_ASLI[NAMA_KOLOM].max()
      x += 1

def normalisasiCost(DATAFRAME_KOSONG,NAMA_KOLOM,DATAFRAME_ASLI):
  panjangKolom = len(df[NAMA_KOLOM])
  x=0
  while x < panjangKolom:
      DATAFRAME_KOSONG.at[x, NAMA_KOLOM] = DATAFRAME_ASLI[NAMA_KOLOM].min()/DATAFRAME_ASLI.at[x, NAMA_KOLOM]
      x += 1

def presentasiBobotWSM(DATAFRAME_KOSONG,NAMA_KOLOM,BOBOT,DATAFRAME_TERNORMALISASI):
  panjangKolom = len(df[NAMA_KOLOM])
  x=0
  while x < panjangKolom:
      DATAFRAME_KOSONG.at[x, NAMA_KOLOM]= DATAFRAME_TERNORMALISASI.at[x, NAMA_KOLOM] * BOBOT
      x +=1

def presentasiBobotWPM(DATAFRAME_KOSONG,NAMA_KOLOM,BOBOT,DATAFRAME_TERNORMALISASI):
  panjangKolom = len(df[NAMA_KOLOM])
  x=0
  while x < panjangKolom:
      DATAFRAME_KOSONG.at[x, NAMA_KOLOM]= DATAFRAME_TERNORMALISASI.at[x, NAMA_KOLOM] ** BOBOT
      x +=1


normalisasiBenefit(kosong,'c1',df)
normalisasiCost(kosong,'c2',df)
normalisasiBenefit(kosong,'c3',df)
normalisasiCost(kosong,'c4',df)
normalisasiBenefit(kosong,'c5',df)

dataTerbobot = pd.DataFrame(columns=df.columns)
dataTerbobotWPM = pd.DataFrame(columns=df.columns)

presentasiBobotWSM(dataTerbobot,'c1',0.3,kosong)
presentasiBobotWSM(dataTerbobot,'c2',0.2,kosong)
presentasiBobotWSM(dataTerbobot,'c3',0.2,kosong)
presentasiBobotWSM(dataTerbobot,'c4',0.2,kosong)
presentasiBobotWSM(dataTerbobot,'c5',0.1,kosong)

presentasiBobotWPM(dataTerbobotWPM,'c1',0.3,kosong)
presentasiBobotWPM(dataTerbobotWPM,'c2',0.2,kosong)
presentasiBobotWPM(dataTerbobotWPM,'c3',0.2,kosong)
presentasiBobotWPM(dataTerbobotWPM,'c4',0.2,kosong)
presentasiBobotWPM(dataTerbobotWPM,'c5',0.1,kosong)

kosong.head()
```
